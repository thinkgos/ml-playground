## 基本元素 - 图片

opencv上的彩色图是以**B-G-R**格式存储的.
图像坐标的起始点`(0, 0)`是在**左上角**, 所以行对应的是y, 向下增长, 列对应的是x, 向右增长
在opencv中, shape为`(y, x, z)`, 其中y为图像的高度, x为图像的宽度, z为图像的通道数.

```python
import cv2
# 
img = cv2.imread('assets/dog.jpeg', cv2.IMREAD_COLOR)
img.shape
# cv2.imshow('lena', img)
# cv2.waitKey(10)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite('assets/output/dog_gray.jpg', img)
```

## 图像基本操作

```python
import cv2

img = cv2.imread('assets/dog.jpeg', cv2.IMREAD_COLOR)

# img[:, :, 0] # 蓝色通道
img[100, 2] # (100,2) 的象素值

# ROI 感兴趣区域
roi = img[100:220, 120:250]
cv2.imwrite('assets/output/dog_roi.jpg', roi)

# 通道分割与合并, 最好是使用numpy中的索引
b, g, r = cv2.split(img)
img = cv2.merge((b, g, r))
```

## 图像几何变换

旋转, 平移, 缩放, 翻转

```python
import cv2
import numpy as np

img = cv2.imread('assets/dog.jpeg')

#! 缩放
# 按照指定的宽度、高度缩放图片
img1 = cv2.resize(img, (132, 150))
cv2.imwrite('assets/output/dog_resize_shrink1.jpg', img1)

# 按照比例缩放，如 x,y 轴均缩小一倍, interpolation 采用缩放方法, 默认采用插值
img2 = cv2.resize(img, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR)
cv2.imwrite('assets/output/dog_resize_shrink2.jpg', img2)

# 按照比例缩放，如 x,y 轴均放大一倍
img3 = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)
cv2.imwrite('assets/output/dog_resize_zoom.jpg', img3)

#! 翻转
# 参数2: 0 表示垂直翻转, >0 表示水平翻转, <0 表示垂直水平翻转
img4 = cv2.flip(img, 1)
cv2.imwrite('assets/output/dog_flip_horizontal.jpg', img4)


#! 平移
rows, cols = img.shape[:2]
# 定义平移矩阵，需要是 numpy 的 float32 类型
# x 轴平移 100，y 轴平移 50
M = np.float32([[1, 0, 100], [0, 1, 50]])

# 用仿射变换实现平移
dst = cv2.warpAffine(img, M, (cols, rows))
cv2.imwrite('assets/output/dog_shift.jpg', dst)

#! 旋转
# 同平移一样, 也是用仿射变换实现的. 因此需要定义一个变换矩阵.

# 参数1: 图片的旋转中心
# 参数2: 旋转角度(正: 逆时针, 负: 顺时针)
# 参数3: 缩放比例, 0.5 表示缩小一半
M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 0.5)
dst = cv2.warpAffine(img, M, (cols, rows))
cv2.imwrite('assets/output/dog_rotate.jpg', dst)
```

## 绘图



```python
import cv2
import numpy as np

img_org = np.zeros((300, 512, 3), np.uint8)

#! 画线
img = img_org.copy()
# 只需要定义起始点和终止点, 以及线的颜色和线宽即可.
cv2.line(img, (0, 0), (512, 300), (255, 0, 0), 5)
cv2.imwrite('assets/output/draw_line.jpg', img)

#! 画矩形
img = img_org.copy()
# 只需要定义矩形的左上角和右下角, 以及线的颜色和线宽即可.
cv2.rectangle(img, (50, 100), (400, 200), (0, 255, 0), 5)
cv2.imwrite('assets/output/draw_rect.jpg', img)

#! 画圆
img = img_org.copy()
# 只需要定义圆的中心和半径, 以及线的颜色和线宽即可. 这里的线宽为-1表示填充圆.
cv2.circle(img, (256, 150), 100, (0, 0, 255), -1)
cv2.imwrite('assets/output/draw_circle.jpg', img)

#! 画椭圆
img = img_org.copy()
# 只需要定义椭圆的中心(x,y), x/y轴的长度, 以及angle旋转角度, startAngle椭圆的起始角度和endAngle椭圆的结束角度, 以及线的颜色和线宽即可.
cv2.ellipse(img, (256, 150), (200, 100), 0, 0, 360, (255, 255, 0), -1)
cv2.imwrite('assets/output/draw_ellipse.jpg', img)

#! 画多边形
img = img_org.copy()
# 定义四个顶点坐标
pts = np.array([[10, 50],  [20, 100], [230, 230], [250, 40]], np.int32)
# 顶点个数：4，矩阵变成 4*1*2 维
# 第一个表示点的数量, 第二个表示通道, 第三个表示坐标
pts = pts.reshape((-1, 1, 2))
# 如果画多条直线, 可以一次性传多个2*1*2维的数组
cv2.polylines(img, [pts], True, (0, 255, 255), 5)
cv2.imwrite('assets/output/draw_polygon.jpg', img)

#! 添加文字
img = img_org.copy()
cv2.putText(img,
            "OpenCv", # 要增加的文本
            (50, 250), # 文字的起始位置(左下角为起始点)
            cv2.FONT_HERSHEY_SIMPLEX, # 字体
            2, # 文字大小
            (255, 255, 255), 
            lineType=cv2.LINE_AA,
            )
cv2.imwrite('assets/output/draw_text.jpg', img)
```
